```python
def cal_zcr(path):
    X,sample_rate = librosa.load(path,res_type='kaiser_fast')
    zcr = np.mean(librosa.feature.zero_crossing_rate(X,frame_length=2048,hop_length=512,center=True))
    return zcr
```

```python
cal_zcr('someAudio/A19DL_09910_1332_26_08_2019.MP3')
```
